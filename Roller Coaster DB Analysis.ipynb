<h1>Cleaning up data

```python
data = data[['coaster_name',
      'Length', 'Speed', 'Location', 'Status',
      #'Opening date',
      #'Type',
    'Manufacturer', 'Height restriction',
      #'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
      #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
      #'height_value', 'height_unit',
      'height_ft',
       'Inversions_clean', 'Gforce_clean']]
```

```python
data = data.rename(columns={
    "coaster_name":"Coaster_Name",
    "Height restriction":"Height_Restriction",
    "year_introduced":"Year_Introduced",
    "opening_date_clean":"Opening_Date",
    "Inversions_clean":"Inversions",
    "Gforce_clean":"Gforce"
    
})
```

```python
data["Opening_Date"] = pd.to_datetime(data["Opening_Date"],format="%Y-%m-%d")
data["Year_Introduced"] = pd.to_datetime(data["Year_Introduced"],format="%Y").dt.year
```

```python
data.tail()
```

```python
data.isna().sum()
```

```python
data[["Coaster_Name","Location","Opening_Date"]].duplicated()
```

```python
data["Height_Restriction"].value_counts()
```

```python
data["Height_Restriction"]=data["Height_Restriction"].str.extract(r"(\d{3})")
```

```python
data["Height_Restriction"]=data["Height_Restriction"].fillna(0)
```

```python
data[["Coaster_Name","Location","Opening_Date"]].duplicated().sum()
```

```python
data= data.drop_duplicates(subset=["Coaster_Name","Location","Opening_Date"],keep="first")\
    .reset_index(drop=True).copy()
```

<h1>What are the top 10 roller coasters in Gforce?<a class = "anchor" id="Q1"></a>

```python
top10_gforce = data.groupby("Coaster_Name")["Gforce"].max().dropna(how="any").sort_values()[-10:]
```

```python
plt.figure(figsize=(12,9))

plt.barh(top10_gforce.index,top10_gforce,color="red")


plt.title("Top 10 coasters with most Gforce")
plt.xlabel("Gforce")
plt.ylabel("Coaster Name")

plt.show()
```

<h1>What are the top 10 roller fastest roller coasters?<a class = "anchor" id ="Q2"></a>

```python
top10_speed = data.groupby("Coaster_Name")["speed_mph"].max().dropna(how="any").sort_values()[-10:]
```

```python
plt.figure(figsize=(10,7))

plt.bar(top10_speed.index,top10_speed,color="blue")


plt.title("Top 10 fastest roller coasters")
plt.xticks(rotation="vertical")
plt.ylabel("Speed in mph")
plt.xlabel("Coaster Name")

plt.show()
```

<h1>Which country has the most roller coasters?<a class = "anchor" id = "Q5"></a>

```python
data.query('Location != "Other"') \
    .groupby('Location')['Coaster_Name'].count()\
    .sort_values()[-10:]\
    .plot(kind = "barh" ,color = "red", title ="Locations with most coasters")
plt.xticks(np.arange(data.Location.value_counts()[1]+1))
plt.show()
```

<h1>which location has the fastest coasters?<a class = "anchor" id = "Q6"></a>

```python
data.query('Location != "Other"')\
    .groupby("Location")["speed_mph"].agg(['mean','count'])\
    .dropna(how="any")\
    .query('count >=10')\
    .sort_values("mean").iloc[-10:,0]\
    .plot(kind = "barh")
plt.show()
```
